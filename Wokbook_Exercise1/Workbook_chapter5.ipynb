Investigate on the difference between is and == given the following variables.

```python
a = 2
b = 2.0
c = a
```

```python
print(a == b)  # True because the values of a and b are the same (both represent the number 2)
print(a is b)  # False because a and b are different variables pointing to different memory locations
print(a == c)  # True because a and c have the same value (both represent the number 2)
print(a is c)  # True because a and c are the same variable pointing to the same memory location
```

Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

Enter a number: 4

Enter a number: 5

Enter a number: bad data

Invalid input

Enter a number: 7

Enter a number: done

16 3 5.333333333333333

```python
total = 0
count = 0

while True:
    try:
        user_input = input("Enter a number: ")
        if user_input == 'done':
            break
        num = int(user_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")

if count > 0:
    average = total / count
    print(total, count, average)
else:
    print("No numbers entered.")
```

Challenge 2

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
numbers = []

while True:
    try:
        user_input = input("Enter a number: ")
        if user_input == 'done':
            break
        num = float(user_input)
        numbers.append(num)
    except ValueError:
        print("Invalid input")

if numbers:
    maximum = max(numbers)
    minimum = min(numbers)
    print("Maximum:", maximum)
    print("Minimum:", minimum)
else:
    print("No numbers entered.")
```
