A. Create an ndarry object (A) with any values from 1 to 100 in a nested Python List. You can choose your range from the given values (1-100).

Given that A is an ndarray object, many functionalities in numpy could be accessed by:
Make sure to explain each line of codes that you have created.
1. Using a property of the ndarray object
    e.g. A.shape
2. Using a method of the ndarray object
    e.g. A.max()
3. Using a numpy function and passing an ndarray
    e.g. np.max(A)

```python
import numpy as np

nested_list = [[i for i in range(1, 10)] for i in range(9)]

A = np.array(nested_list)

#1. Using a property of the ndarray object
print("Shape of A:", A.shape) #A.shape returns a tuple with each index having the number of corresponding elements. It indicates that A has 9 rows and 9 columns.

#2. Using a method of the ndarray object
max_value = A.max() #A.max() is a method that returns the maximum value in the ndarray A.
print("Maximum value in A:", max_value)  # To display the maximum number which is 9.

#3. Using a numpy function and passing an ndarray
min_value = np.min(A) #np.min(A) is a numpy function that returns the minimum value in the ndarray A.
print("Minimum value in A:", min_value) #To display the minimum number which is 1.
```

B. Create a 5x5 integer-array filled with ones.

Check its shape, size, and number of dimensions. Explain briefly the codes you have created by writing a comment beside the code.

```python
import numpy as np

ones_array = np.ones((5, 5), dtype=int)

print("Shape of the array:", ones_array.shape) #The shape property of the array returns a tuple representing the dimensions of the array. It's (5, 5) indicating the array has 5 rows and 5 columns.

print("Size of the array:", ones_array.size) #The size property of the array returns the total number of elements in the array. For a 5x5 array, it will be 5*5 = 25.

print("Number of dimensions:", ones_array.ndim) #The ndim property of the array returns the number of dimensions (axes) of the array. It displays 2.

```

C. Create an array of values from -5 to + 5. Multiply each value with 2. Print the resulting array.

```python
import numpy as np

original_array = np.arange(-5, 6)

resulting_array = original_array * 2 # This line of code multiply each value by 2.

print("Resulting array:", resulting_array) # Print the resulting array.
```

A. Create a four-dimensional array of random integers in the interval [0, 12]. Print the result.

Get the dimensions, shape, and size.
Explain what is the result briefly by writing a comment before the code.

```python
import numpy as np

# I created a four-dimensional array of random integers in the interval [0, 12].
four_dim_array = np.random.randint(0, 13, size=(3, 4, 2, 5))

# Print dimensions, shape, and size of the array
print("Dimensions of the array:", four_dim_array.ndim)
print("Shape of the array:", four_dim_array.shape)
print("Size of the array:", four_dim_array.size)
```

B. Get the maximum number for each dimensions and display it.

Explain what is the result briefly by writing a comment before the code.

```python
# Getting the maximum number for each dimension and display the maximum number which is 12.
max_per_dimension = np.amax(four_dim_array, axis=(0, 1, 2, 3))
print("Maximum number for each dimension:", max_per_dimension)
```
